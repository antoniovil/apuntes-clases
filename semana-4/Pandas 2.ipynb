```python
def limpiar_fecha(x):
    fecha_string = x.split(' ')[0]
    fecha_date = datetime.datetime.strptime(fecha_string, '%Y-%m-%d')
    return fecha_date
```

```python
dff['fecha'] = dff['invoicedate'].apply(limpiar_fecha)
dff.head(2)
```

```python
from geopy.geocoders import Nominatim
```

```python
#ejecutamos funcion

y = 'United Kingdom'
geolocator = Nominatim(user_agent = 'antonio')
location = geolocator.geocode(y)
```

```python
location
```

```python
location.longitude
```

```python
def extraer_coordenadas(x):
    try:
        geolocator = Nominatim(user_agent = 'antonio')
        location = geolocator.geocode(x)
        return location.latitude, location.longitude
    except:
        return 'no hay datos'
```

```python
dff['coordernadas'] = dff['country'].apply(extraer_coordenadas)
dff.head(1)
```

```python
#apply a multiple columnas

def precio_total(x,y):
    return x*y
```

```python
#apply con lambda para cuando hay mas d un parametro

dff['precio_total'] = dff.apply(lambda data: precio_total(data['quantity'], data['unitprice']), axis = 1)
dff.head(2)
```

```python
#GROUPBY 

dff.head(2)
```

```python
#resetear indices
dff.reset_index(inplace = True)
dff.head(2)
```
