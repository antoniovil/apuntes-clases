```python
import pandas as pd
import numpy as np
from pymongo import MongoClient
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.max_columns = None
```

```python
client = MongoClient()
client.list_database_names()
mydb = client['ironhack']
sandwich = mydb.Sandwich

houses = pd.DataFrame(sandwich.find())
houses.head()
```

```python
columns_to_float = ['delivery_charges',
       'customer_lat', 'customer_long','order_total',
       'distance_to_nearest_warehouse']
columns_to_int = [ 'index',
       'order_price', 'coupon_discount',]
```

```python
houses1 = houses[columns_to_float].astype(dtype = float)
houses2 = houses[columns_to_int].astype(dtype = int)
print(houses2.dtypes)
print(houses1.dtypes)
```

```python
houses[columns_to_float] = houses1
houses[columns_to_int] = houses2
houses.head(2)
```

```python
houses.set_index(["_id"], inplace = True)
```

```python
houses['season'] = houses['season'].str.lower()
```

```python
houses.duplicated(['order_id']).sum()
houses.drop_duplicates(['order_id'], inplace = True)
```

```python
ventas = pd.DataFrame(houses.groupby('season')['order_total'].sum())
ventas.reset_index(inplace = True)
ventas.head(4)
```

```python
ax = plt.axes() # para poder cambiar el estilo de los ejes

ax.set_facecolor("white") # poner el fondo en blanco

ax.spines['bottom'].set_color('black') # color del eje x
ax.spines['top'].set_color('white') # color del eje de arriba
ax.spines['right'].set_color('white') # color del eje de la derecha
ax.spines['left'].set_color('black') # color del eje y

plt.plot(ventas['season'], # eje x
         ventas['order_total'], # eje y
         color = "black", # para cambiar el color de la línea
         linewidth = 2, # para cambiar el grosor de la línea
         marker = "o") # para poner los marcadores en nuestra línea


plt.show();
```
